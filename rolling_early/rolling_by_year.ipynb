```python
import rubin_sim
from rubin_sim import maf
%matplotlib inline
import healpy as hp
import numpy as np
```

```python
baseline_file = 'roll_early_v2.0_10yrs.db'
runname = baseline_file.split('/')[-1].replace('.db', '')
```

```python
slicer = maf.HealpixSlicer()


# just count how many observations overlap the point
metric = maf.CountMetric(col='night')
bundle_list = []

for night in np.arange(365.25, 365.25*11, 365.25):
# limit the data to just things on one night
    sql='night<%i and note not like "%%DD%%"' % night

    bundle_list.append(maf.MetricBundle(metric, slicer, sql))

bg = maf.MetricBundleGroup(bundle_list, baseline_file, '.', None)
```

```python
bg.runAll()
bg.plotAll(closefigs=False)
```
